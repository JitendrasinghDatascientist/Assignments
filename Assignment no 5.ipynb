```python
def is_palindrome(s):
    s = s.lower()  # Convert to lowercase for case-insensitive comparison
    s = ''.join(filter(str.isalnum, s))  # Remove non-alphanumeric characters
    
    return s == s[::-1]  # Compare the string with its reverse

# Test the function
input_string = "A man, a plan, a canal, Panama!"
result = is_palindrome(input_string)
if result:
    print(f"'{input_string}' is a palindrome.")
else:
    print(f"'{input_string}' is not a palindrome.")
```

```python
def remove_duplicates(input_list):
    unique_list = []
    for item in input_list:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list
input_list = [1, 2, 2, 3, 4, 4, 5, 6, 6, 7]
result = remove_duplicates(input_list)
print(result)
```

NumPy and pandas are two fundamental libraries in the Python ecosystem for data manipulation and analysis. While they have some overlapping functionalities, they serve different purposes and have distinct features. Here are five key differences between NumPy and pandas:

Purpose and Focus:

NumPy (Numerical Python): NumPy is primarily focused on providing support for multidimensional arrays and matrices, along with mathematical functions to operate on these arrays efficiently. It's used for numerical computations and data manipulation at a low level.
pandas: pandas, on the other hand, is designed to work with structured data and provides high-level data structures like DataFrames and Series. It's optimized for data analysis, cleaning, and manipulation.
Data Structures:

NumPy: The core data structure in NumPy is the ndarray (n-dimensional array), which is a homogeneous array of elements with a fixed data type.
pandas: pandas introduces two primary data structures: Series (1D labeled array) and DataFrame (2D labeled table). These structures can hold heterogeneous data types.
Functionality:

NumPy: Provides a wide range of mathematical and array operations, including element-wise operations, linear algebra, Fourier transforms, and more.
pandas: Offers data manipulation, cleaning, merging, reshaping, and analysis tools. It integrates data from various sources and provides powerful data grouping and aggregation capabilities.
Indexing and Labeling:

NumPy: Uses integer-based indexing to access elements in arrays.
pandas: Utilizes labels and indexes to access data, providing more flexibility and readability. DataFrame columns and rows can be labeled, making it easier to work with real-world data.
Domain and Use Cases:

NumPy: Used primarily for mathematical and scientific computations, such as numerical simulations, signal processing, and linear algebra operations.
pandas: Suited for data analysis and manipulation tasks, such as data cleaning, exploration, transformation, and visualization. It's commonly used in fields like data science, finance, and social sciences.
